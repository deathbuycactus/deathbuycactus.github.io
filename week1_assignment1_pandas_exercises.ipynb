##### 6. Find the total cargo transported (mail + freight) and mean distance traveled for the 10 carriers that transported the most cargo.

```python
mail_by_carrier = flights.groupby('carrier_name')['mail'].sum().reset_index()
freight_by_carrier = flights.groupby('carrier_name')['freight'].sum().reset_index()
cargo_by_carrier = pd.merge(mail_by_carrier, freight_by_carrier, on='carrier_name', how='outer')
cargo_by_carrier['cargo'] = cargo_by_carrier['mail'] + cargo_by_carrier['freight']
cargo_by_carrier = cargo_by_carrier.sort_values(by='cargo', ascending = False)
cargo_by_carrier.head(10)
```
